* Create a 4 row and 1 column plot where:
    * The first row is a line graph of all the first quater earnings
    * The second row is a line graph of all the second quater earnings
    * The third row is a line graph of all the third quater earnings
    * The fourth row is a line graph of all the fourth quater earnings

* The y-axis should be called "Earnings"

````python
# Your code below (You need to make the appropriate imports of matplotlib and numpy)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline

text = '''
Index Qtr1  Qtr2  Qtr3  Qtr4
----  ----  ----  ----  ----
1960  0.71  0.63  0.85  0.44
1961  0.61  0.69  0.92  0.55
1962  0.72  0.77  0.92  0.60
1963  0.83  0.80  1.00  0.77
1964  0.92  1.00  1.24  1.00
1965  1.16  1.30  1.45  1.25
1966  1.26  1.38  1.86  1.56
1967  1.53  1.59  1.83  1.86
1968  1.53  2.07  2.34  2.25
1969  2.16  2.43  2.70  2.25
1970  2.79  3.42  3.69  3.60
1971  3.60  4.32  4.32  4.05
1972  4.86  5.04  5.04  4.41
1973  5.58  5.85  6.57  5.31
1974  6.03  6.39  6.93  5.85
1975  6.93  7.74  7.83  6.12
1976  7.74  8.91  8.28  6.84
1977  9.54 10.26  9.54  8.73
1978 11.88 12.06 12.15  8.91
1979 14.04 12.96 14.85  9.99
1980 16.20 14.67 16.02 11.61
'''
def to_float(lst):
    return [float(word) if word.count(".") > 0 else word for word in lst]

def clean_text(textlist):
    newlines = list()
    words = list()
    column_names = list()
    for line in textlist:
        if line.count("Index") > 0:
            column_names = line.split(" ")
        if line != "" and line.count("-")==0 and line.count("Index") == 0:
            newwords = list()
            words = line.split(" ")
            for word in words:
                if word != "":
                    newwords.append(word)
            newlines.append(newwords)
    return newlines, column_names

# Text_lines has lines but for '\n', including column names and ----- ----- line
text_lines = text.split('\n')
#print(text_lines)

# Each row is a date and the data, all str() fields
lines_words, col_names = clean_text(text_lines)
#print(lines_words[:])

# Extract column names: Index, Qtr1-4
c_names = [name for name in col_names if name != ""]
#print(c_names)

( """
Set lines_wordsnp to easily process fields for further cleaning. Still includes
column names and row names/indexes. Then generate pandas DataFrame. Test print by column name.
""")
lines_wordsnp = np.array(lines_words)
#print(lines_wordsnp)
lines_wordspd = pd.DataFrame(index=lines_wordsnp[0::,0].astype("int64"), data=lines_wordsnp[0::,1::], 
                             columns=c_names[1::])

years = lines_wordspd.index
#print(years)
#print(lines_wordspd["Qtr1"]) # Test prints ok

# Didn't need for Challenge 1
#years = np.array([float(index)+y for y in np.arange(0,1,.25) for index in years]).astype("float64")
#years.sort()

#years[years%2==0].tolist()
````

````python
fig, axs = plt.subplots(4,1, constrained_layout=True, figsize=(8,8), sharey=False, sharex="all",
                       squeeze=False)

xlabs = ["","","","Year"]
ylabs = ["Earnings","Earnings","Earngings","Earnings"]
x_ticks = [[],[],[],years[years%2==0]]
flat_data = np.array(lines_wordspd).astype("float64")
xlims = [1959,1982]
ylims = [0,20]
y_ticks = np.arange(-1,20,5)

for i, ax in enumerate(axs.flatten()):
    ax.plot(years, lines_wordspd.iloc[::,i])
    ax.set_title(lines_wordspd.keys()[i])
    ax.set_xlabel(xlabs[i])
    ax.set_xticks(x_ticks[i])
    ax.set_yticks(y_ticks)
    ax.axis([1959, 1982, -1, 20])
    plt.ylim(bottom=-1,top=20)
    plt.xlim(left=1959,right=1982)
    
fig.suptitle("Johnson and Johnson Quarterly Earnings")
plt.show()
````
